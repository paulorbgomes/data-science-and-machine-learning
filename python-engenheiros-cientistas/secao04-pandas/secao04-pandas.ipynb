- **ESTRUTURA DE DADOS NO PANDAS:**

* **Séries:** são estruturas de dados 1D e podem ser criadas a partir de dicionários, arrays, entre outros. É muito parecido com os arrays em termos de propriedades e métodos.
```python
pd.Series(data, index=index)
```

* **DataFrame:** são estruturas de dados 2D e podem ser criados a partir de dicionários, dados estruturados em 2D, arquivos csv, etc. Em resumo, Dataframe é uma tabela.
```python
pd.DataFrame({'key_1': pd.Series(...), 'key_n': pd.Series(...)})
pd.read_csv(file)
```

- **FUNÇÕES MERGE | CONCAT: cruzando dados de tabelas diferentes**

  * **Referência:** *https://pandas.pydata.org/docs/user_guide/merging.html*
  * Função **concat** auxilia em combinar, concatenar os DataFrames. A função **merge** ajuda a fundir os DataFrames, tendo uma funcionalidade parecida com o procv do Excel.
  * pd.concat(objs: Union[Iterable[FrameOrSeries], Mapping[Label, FrameOrSeries]], axis='0', join: str="outer", ignore_index: bool='False', keys='None', levels='None', names='None', verify_integrity: bool='False', sort: bool: 'False', copy: bool='True')
  * pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=True, suffixes=('_x','_y'), copy=True, indicator=False, validate=None)

**pd.concat(args)**

````python
import pandas as pd
import numpy as np

dfa = pd.DataFrame(np.arange(15).reshape(5,3))
dfb = pd.DataFrame((np.arange(9)**3).reshape(3,3))
pd.concat([dfa,dfb],ignore_index=True)
````

**pd.merge(args)**

````python
# Caso com colunas equivalentes mas com os mesmos nomes ...
df1 = pd.DataFrame({"id":[1,2,3],
                    "cor":["branco","preto","verde"]})
df2 = pd.DataFrame({"times":["time1","time2","time3","time4","time5"],
                    "id":[1,1,2,3,2]})
pd.merge(df1,df2)
````

````python
# Caso com colunas equivalentes mas com os nomes diferentes ...
df1 = pd.DataFrame({"id":[1,2,3],
                    "cor":["branco","preto","verde"]})
df2 = pd.DataFrame({"times":["time1","time2","time3","time4","time5"],
                    "id_cor":[1,1,2,3,2]})
pd.merge(df1,df2,left_on="id",right_on="id_cor")
````

- **PRÉ-PROCESSAMENTO: LIMPEZA DOS DADOS**
  * .drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise'): remove uma série de dados especificada;
  * .isnull(obj) | .notnull(obj): cria uma série de booleanos;
  * .dropna(axis=0, how='any', thresh=None, subset=None, inplace=False): deleta linha (axis=0) ou coluna (axis=1) com cédula(s) nula(s);
  * .fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None): substitui o valor nulo por um valor determinado;
  * .duplicated(subset=None, keep='first'): retorna um booleano com valores duplicados;
  * .drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False): deleta linhas com valores duplicados. Pode selecionar uma determinada coluna usando subset.
  * **Nota: esses métodos não modificam o DataFrame.**

````python
import pandas as pd
import numpy as np

x = np.arange(20).reshape(4,5)
df_columns = ["A","B","C","D","E"]
df = pd.DataFrame(x,columns=df_columns)
df
````

````python
df.drop(index=0)     # remoção de linha
df.drop(columns="A") # remoção de coluna
df.drop(index=2,columns="C")
````

````python
b = np.vstack((x, np.array([np.nan,np.nan,np.nan,np.nan,np.nan])))
df = pd.DataFrame(b,columns=df_columns)
df.isnull()
df.notnull()
df[df["A"].notnull()]
````

````python
df.dropna(axis=0)   # axis=0 | axis=1
````

````python
c = np.vstack((x, np.array([1,np.nan,np.nan,np.nan,5])))
df = pd.DataFrame(c,columns=df_columns)
df.dropna(how="all")   # how="all" | how="any"
df.dropna(axis=1,how="any")   # how="all" | how="any"
````

````python
df.fillna(value=99)
````

````python
d = np.vstack((x, np.array([15,np.nan,np.nan,np.nan,np.nan])))
df = pd.DataFrame(d,columns=df_columns)
df.duplicated()
df.duplicated(subset="A")
````

````python
df.drop_duplicates(subset="A")
````

- **APLICAÇÃO: REGRESSÃO LINEAR SIMPLES COM STATSMODELS**

````python
# Bibliotecas
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# Dataset ...
df = pd.DataFrame({"X":[0,1,2,3,4,5,6,7],
                  "Y":[2,3,5,10,15,20,24,30]})

# Visualização do dataset ...
X = df.iloc[:,0]
Y = df.iloc[:,1]
plt.figure()
plt.scatter(X,Y,color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regressão Linear Simples")
plt.show()
````

````python
# Regressão linear simples com statsmodels
X1 = sm.add_constant(X)      # adiciona a constante de regressão linear, se desejável
result = sm.OLS(Y,X1).fit()  # gera o objeto de regressão linear baseado em X e Y
result.params                # retorna os parâmetros da regressão linear
print(result.summary())      # imprime um relatório estatístico da regressão linear
````

````python
Yhat = result.params[1]*X + result.params[0]
plt.scatter(X,Y,color="red",label="Dados Originais")
plt.plot(X,Yhat,"k--",label="Modelo")
#plt.plot(X,result.predict(),"k--",label="Modelo")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regressão Linear Simples [StatsModels]")
````

- **APLICAÇÃO: REGRESSÃO LINEAR MÚLTIPLA COM STATSMODELS**

````python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline

file_path = "datasets_1358_30676_Players.csv"
df = pd.read_csv(file_path)
df.head()

# Tratamento dos dados
df = df.dropna(axis=0)
peso = df["weight"]
altura = df["height"]
idade = 2024 - df["born"]
df["age"] = idade
df.columns
````

````python
# Análise 1 (Regressão Linear Simples)
X1 = sm.add_constant(altura)
result = sm.OLS(peso,X1).fit()
Yhat = result.params[1]*altura + result.params[0]
plt.figure()
plt.scatter(altura,peso,color="blue")
#plt.plot(altura,Yhat,"r--")
plt.plot(altura,result.predict(),"r--")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Weight vs. Height")
plt.grid(True)
plt.show()
print(f"Coeficiente angular: {result.params[1]}; Coeficiente linear: {result.params[0]}")
````

````python
# Análise 2 (Regressão Linear Simples)
X1 = sm.add_constant(idade)
result = sm.OLS(peso,X1).fit()
Yhat = result.params[1]*idade + result.params[0]
plt.figure()
plt.scatter(idade,peso,color="blue")
plt.plot(idade,result.predict(),"r--")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight vs. Age")
plt.grid(True)
plt.show()
print(f"Coeficiente angular: {result.params[1]}; Coeficiente linear: {result.params[0]}")
````
