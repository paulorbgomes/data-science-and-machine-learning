# BIBLIOTECA MATPLOTLIB

- **PYPLOT E ELEMENTOS GRÁFICOS:**

  * **Referência:** *https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html*

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Método 1:
x = np.linspace(0,2*np.pi,500)
y = np.sin(x)
z = np.cos(x)

# Geração do gráfico
plt.plot(x,y,"k-",label="sin(x)")
plt.plot(x,z,"r--",label="cos(x)")
plt.xlabel("Eixo-x")
plt.ylabel("Eixo-y")
plt.title("Funções sin(x) e cos(x)")
plt.grid(True)
plt.legend()
````

````python
# Método 2:
fig, axes = plt.subplots()
axes.plot(x,y,"k-",label="sin(x)")
axes.plot(x,z,"r--",label="cos(x)")
axes.grid(True)
axes.set_title("Funções sin(x) e cos(x)")
axes.set_xlabel("Eixo-x")
axes.set_ylabel("Eixo-y")
plt.legend()
````

- **SUBPLOTS:**

  * "Vários gráficos em uma mesma janela;"

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Método 1:
x = np.linspace(0,4*np.pi,500)
y1 = np.cos(x)
y2 = np.sin(x)

y3 = np.exp(x)
y3 = (1 / y3.max()) * y3

y4 = np.exp(-x)
y4 = (1 / y4.max()) * y4

fig, (ax1,ax2) = plt.subplots(2,2)
ax1[0].plot(x,y1,"k--")
ax1[1].plot(x,y2,"r--")
ax2[0].plot(x,y3,"g--")
ax2[1].plot(x,y4,"b--")
ax1[0].grid(True)
ax1[1].grid(True)
ax2[0].grid(True)
ax2[1].grid(True)
````

````python
# Método 2:
plt.figure()
plt.subplot(221)
plt.plot(x,y1,"k--")
plt.plot(x,y2,"r--")
plt.grid(True)
plt.subplot(222)
plt.plot(x,y2,"r--")
plt.grid(True)
plt.subplot(223)
plt.plot(x,y3,"g--")
plt.grid(True)
plt.subplot(224)
plt.plot(x,y4,"b--")
plt.plot(x,x,"k--")
plt.grid(True)
````

- **ESCALA LOG:**
  ```python
  plt.yscale("log")
  ```

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,10,500)
y = np.exp(x)

plt.figure()
plt.suptitle("Comparação Escalas")
plt.subplot(211)
plt.plot(x,y,"r--",label="Escala Linear")
plt.ylabel("Y [linear]")
plt.grid(True)
plt.legend()
plt.subplot(212)
plt.plot(x,y,"b--",label="Escala Logarítmica")
plt.ylabel("Y [log]")
plt.yscale("log")
plt.grid(True)
plt.legend()
plt.show()
````

- **EIXO SECUNDÁRIO:**

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,10,100)
y = np.exp(x)

fig, ax1 = plt.subplots()
ax1.plot(x,y,"r--")
ax1.set_xlabel("X")
ax1.set_ylabel("Y1",color="r")
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(x,y,"g--")
ax2.set_ylabel("Y2",color="g")
plt.yscale("log")
plt.show()
````

- **GRÁFICO DE BARRAS:**

  **Referência:** *https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html*
  ```python
  matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, , align="center", data=None. *kwargs)
  ```

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

y = [20, 50, 10, 30, 5]
x = list(range(1,len(y)+1))
plt.figure()
plt.bar(x,y,label="Legenda")
plt.xlabel("Eixo-x")
plt.ylabel("Eixo-y")
plt.legend()
plt.show()
````

````python
x = ["A", "B", "C", "D", "E"]
y2 = [10,20,30,20,10]

plt.figure()
plt.bar(x,y,label="Legenda 1",color="gray")
plt.bar(x,y2,label="Legenda 2", color="black")
plt.xlabel("Eixo-x")
plt.ylabel("Eixo-y")
plt.legend()
plt.show()
````

````python
fig, ax = plt.subplots()

x = np.arange(5)
largura = 0.4

rect1 = ax.bar(x+largura/2,y,color="gray",width=largura,label="Valores 1")
rect2 = ax.bar(x-largura/2,y2,color="black",width=largura, label="Valores 2")
ax.set_xticks(x)
ax.set_xticklabels(["A", "B", "C", "D", "E"])
plt.legend()
plt.show()
````

- **GRÁFICO DE PIZZA:**

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = [10,20,30,20,10,7,3]
nomes = ["A","B","C","D","E","F","G"]
explodir = (0,0,0.2,0,0,0,0.2)

plt.figure()
plt.pie(x,labels=nomes,autopct="%1.1f%%",explode=explodir,shadow=False)
plt.show()
````
