# Objeto Series

```python
import pandas as pd

# Sintaxe:
# pd.Series([lista_valores], index=[lista_indices], name="Nome para o objeto")

lista = ["Morango", "Maçã", "Uva", "Pêssego", "Carambola", "Lichia"]
s1 = pd.Series(lista, name="Frutas")

print(lista)

print(s1)
print(s1.dtype)
print(s1.values)
print(s1.index)

print(s1[2])
```

```python
indices = ["Mo", "Ma", "U", "Pe", "Ca", "Li"]
s1 = pd.Series(lista, index=indices, name="Frutas")
print(s1)
```

```python
print(f"Número de frutas: {s1.count()}")
print(s1.str.upper())
```

```python
lista = [4.003, 20.180, 39.95, 83.798, 131.293, 222]
gasesNobresA = pd.Series(lista)

print(gasesNobresA.mean())
print(gasesNobresA.median())
print(gasesNobresA.min())
print(gasesNobresA.max())
print(gasesNobresA.std())
```

```python
gasesNobresA.describe()
```

```python
for item in gasesNobresA:
  print(item, item*2)

comb = gasesNobresA.map(lambda item: 2*item)
print(comb)
```

# Exemplo prático

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
arquivo = pd.read_csv("VALE3.csv", sep=",")
arquivo
```

```python
df = arquivo[["DATA","FECHAMENTO"]].copy()
df.columns = ["Data","Fechamento"]
df["Fechamento"] = df["Fechamento"].str.replace(",",".").astype(float)
df
```

```python
df.to_csv("VALE3_tratado.csv",index=False)
```

```python
df = pd.read_csv("VALE3_tratado.csv")
df
```

```python
print("Preço médio:")
df["Fechamento"].mean()
```

```python
print("Data e preço mínimo:")
df[df["Fechamento"] == df["Fechamento"].min()]
```

```python
print("Data e preço máximo:")
df[df["Fechamento"] == df["Fechamento"].max()]
```

```python
df_comprado = df["Fechamento"][9]
df_especulado = df["Fechamento"][69]
valor = (df_especulado - df_comprado) * 2000
if valor < 0:
  print(f"Prejuízo de R$ {abs(valor):.2f}")
else:
  print(f"Lucro de R$ {abs(valor):.2f}")
```

```python
df["Data"] = pd.to_datetime(df["Data"], format="%d/%m/%Y")
plt.figure(figsize=(12,6))
plt.plot(df["Data"],df["Fechamento"],label="Preço de Fechamento",color="blue")
plt.title("Preço de Fechamento da VALE3")
plt.xlabel("Período")
plt.ylabel("Preço de Fechamento (R$)")
plt.legend()
plt.grid(True)
plt.show()
```
