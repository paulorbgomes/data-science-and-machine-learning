```python
# Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()

# Dados
data = pd.read_csv("FuelConsumptionCo2.csv")
#data
#data.describe()
y = data["CO2EMISSIONS"]
x1 = data["FUELCONSUMPTION_CITY"]
plt.scatter(x1,y)
plt.xlabel("FUEL CONSUMPTION")
plt.ylabel("CO2 EMISSIONS")
plt.show()
```

```python
# Regress√£o Linear Simples
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
# Plotagem da Regress√£o Linear Simples
plt.scatter(x1,y)
yhat = 13.8764*x1 + 71.7202
fig = plt.plot(x1, yhat, lw=3, c="red", label="Regress√£o Linear Simples")
plt.xlabel("FUEL CONSUMPTION")
plt.ylabel("CO2 EMISSIONS")
plt.show()
```

```python
# Predi√ß√£o
x
new_data = pd.DataFrame({"const":[1], "FUELCONSUMPTION_CITY":[15]})
new_data = new_data[["const", "FUELCONSUMPTION_CITY"]]
new_data
predictions = results.predict(new_data)
predictions
predictionsdf = pd.DataFrame({"Predictions":predictions})
joined = new_data.join(predictionsdf)
joined
```
