```python
# Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Dados
data = pd.read_csv("real_estate_price_size.csv")
data
#data.describe()
```

```python
# VisualizaÃ§Ã£o dos Dados
y = data["price"]
x1 = data["size"]
plt.scatter(x1, y)
plt.xlabel("SIZE", fontsize=20)
plt.ylabel("PRICE", fontsize=20)
plt.show()
```

```python
# RegressÃ£o Linear Simples (utilizando statsmodel)
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
# Plotagem do Modelo de RegressÃ£o Linear Simples
plt.scatter(x1, y)
yhat = 223.1787 * x1 + 1.019e+05
fig = plt.plot(x1, yhat, lw=4, c="orange", label="RegressÃ£o Linear Simples")
plt.xlabel("SIZE")
plt.ylabel("PRICE")
plt.show()
```
