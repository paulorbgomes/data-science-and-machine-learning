## Aplicação da função map() em Series

```python
df_mean = df["points"].mean()
df_std = df["points"].std()
df_p = df["points"].map(lambda i: (i - df_mean) / df_std)     # padronização doo atributo escolhido
df_p
df_p.mean()
df_p.std()
```

```python
def contagem(array):
  array = pd.Series(array)
  tropical = 0
  fruity = 0
  for i in array:
    if("tropical" in i):
      tropical = tropical + 1
    if("fruity" in i):
      fruity = fruity + 1
  return (tropical, fruity)

escriptor_counts = df["description"].apply(contagem, 1)

tropical = 0
fruity = 0
for tupla in escriptor_counts:
  tropical = tropical + tupla[0]
  fruity = fruity + tupla[1]

print(f"Tropical: {tropical}, Fruity: {fruity}")
```

```python
fruity = df["description"].str.contains("fruity").sum()
tropical = df["description"].str.contains("tropical").sum()
print(f"Tropical: {tropical}, Fruity: {fruity}")
```
