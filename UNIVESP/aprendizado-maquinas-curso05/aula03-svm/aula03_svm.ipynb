```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

plt.rcParams["figure.figsize"]=[10,7]
```

```python
url = "social-network-ads.csv"
df = pd.read_csv(url, sep=",")
df.head()
```

```python
df["Age"].plot.hist()
plt.show()
```

```python
plt.figure(figsize=(10,7))
df["EstimatedSalary"].plot.hist()
plt.show()
```

```python
fig, ax = plt.subplots()
colors = {0:"red", 1:"green"}
ax.scatter(df["Age"], df["EstimatedSalary"], c=df["Purchased"].map(colors))
plt.show()
```

```python
atributos = df[["Age","EstimatedSalary"]]
classes = df["Purchased"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(atributos, classes, train_size=0.7, random_state=52)
```

```python
scaler = StandardScaler()
treino = scaler.fit_transform(X_train)
teste = scaler.fit_transform(X_test)
```

```python
model = SVC(kernel="poly", degree=3, gamma="scale", C=1.0, coef0=2)
model.fit(treino, y_train)
```

```python
predicao = model.predict(teste)
acuracia = accuracy_score(y_test,predicao)
acuracia
```

```python
model = SVC(kernel="sigmoid", gamma="scale", C=0.2, coef0=2)
model.fit(treino, y_train)
```

```python
predicao = model.predict(teste)
acuracia = accuracy_score(y_test,predicao)
acuracia
```
