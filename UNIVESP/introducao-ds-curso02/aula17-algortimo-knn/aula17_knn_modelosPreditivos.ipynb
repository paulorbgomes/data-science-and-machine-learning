# ALGORITMO KNN E AVALIAÇÃO DE MODELOS PREDITIVOS

## Importar Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

## Importar dataset iris

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

## Gerar um target (alvo)

```python
df["target"] = iris.target
```

```python
df
```

## Gerar um target name (apenas para demonstrar)

```python
iris.target_names
```

```python
df["target_name"] = iris.target_names[df["target"]]
```

```python
df
```

## Determinar as features (colunas, atributos) que farão parte do estudo

```python
iris_features = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
```

## Determinar o alvo (Y), lembrando que por fazer uma classificação usando KNN precisamos de um alvo numérico.

```python
Y = df["target"]
# Y = df.target
```

## Selecionar os dados, as features que serão utilizadas (X)

```python
X = df[iris_features]
```

```python
X
```

```python
Y
```

## Usar Scikit-Learn
### Define = Escolha de modelo (define parâmetros)
### Fit = Treinar
### Predict = Fazer Predição
### Evaluate = Avaliar os Resultados

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
modelo =  KNeighborsClassifier(15)
```

```python
modelo.fit(X,Y)
```

```python
print(modelo.predict(X))
```

```python
modelo.score(X,Y)
```
