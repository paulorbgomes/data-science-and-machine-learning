  ## Analisando um Conjunto de Dados com o Pandas (Análise Descritiva)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={"figure.figsize":(10,6)})
```

```python
df = pd.read_csv("dataset_educacao_sp.csv", sep=";")
df.head()
```

```python
df.info()
```

```python
filtro = df[df["ANOS INICIAIS"] > 0]
sns.displot(filtro["ANOS INICIAIS"], bins=10)
plt.show()
```

```python
filtro = df[df["ANOS FINAIS"] > 0]
sns.displot(filtro["ANOS FINAIS"], bins=10)
plt.show()
```

```python
filtro = df[df["ENSINO MÉDIO"] > 0]
sns.displot(filtro["ENSINO MÉDIO"], bins=10)
plt.show()
```

```python
fundamental_iniciais = df["ANOS INICIAIS"].sum()
fundamental_finais = df["ANOS FINAIS"].sum()
medio = df["ENSINO MÉDIO"].sum()
total = fundamental_iniciais + fundamental_finais + medio
print(total)
```

```python
alunos = [fundamental_iniciais, fundamental_finais, medio]
periodo = ["Fundamental - Anos Iniciais", "Fundamental - Anos Finais", "Ensino Médio"]
plt.pie(alunos,labels=periodo,autopct="%0.0f%%")
plt.show()
```

```python
filtro = df[df["ENSINO MÉDIO"] > 0]
sns.boxplot(y=filtro["ENSINO MÉDIO"])
plt.show()
```

```python
filtro = df[(df["ANOS INICIAIS"] > 0) & (df["ANOS FINAIS"] > 0)]
sns.scatterplot(x=filtro["ANOS INICIAIS"],y=filtro["ANOS FINAIS"])
plt.show()
```
