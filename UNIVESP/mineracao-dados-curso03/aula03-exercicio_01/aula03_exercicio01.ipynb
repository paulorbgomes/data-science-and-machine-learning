# EXERCÍCIO 01
  ## Limpeza e Redução em Base de Dados

```python
import pandas as pd
```

```python
url = "https://github.com/higoramario/univesp-com360-mineracao-dados/raw/main/dados-covid-limpeza.csv"
df_root = pd.read_csv(url)
df_root.head()
```

```python
df_root.info()
```

```python
df_root.describe()
```

```python
df_root["uf"].value_counts()
```

```python
df_root["uf"] = df_root["uf"].str.upper()
df_root["uf"].value_counts()
```

```python
df_root["renda"].value_counts()
```

```python
df_root["vacina"].value_counts()
```

```python
df_root["idade"] = df_root["idade"].fillna(round(df_root["idade"].mean()))
df_root["idade"].value_counts()
```

```python
df_root["uf"] = df_root["uf"].fillna(method="ffill")
df_root["uf"].value_counts()
```

```python
df_root["renda"] = df_root["renda"].fillna(method="bfill")
df_root["renda"].value_counts()
```
