# AN√ÅLISE DE AGRUPAMENTO (Complete-Linkage)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df_root = pd.read_csv("shopping-data.csv")
df_root.head()
```

```python
df = df_root[["Annual Income (k$)","Spending Score (1-100)"]].copy()
df.head()
```

```python
plt.scatter(df["Annual Income (k$)"],df["Spending Score (1-100)"])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
plt.title("Dendogram")
dendograma = shc.dendrogram(shc.linkage(df, method="complete"))
```

```python
n_grupos = 5
model = AgglomerativeClustering(n_clusters=n_grupos, affinity="euclidean", linkage="complete")
model.fit_predict(df)
```

```python
plt.scatter(df["Annual Income (k$)"],df["Spending Score (1-100)"],c=model.labels_,cmap="rainbow")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
fig,eixo=plt.subplots()
plt.scatter(df["Annual Income (k$)"],df["Spending Score (1-100)"],c=model.labels_,cmap="rainbow")
eixo.set_xlabel("Annual Income (k$)")
eixo.set_ylabel("Spending Score (1-100)")

for i,texto in enumerate(df_root["Genre"]):
  if texto == "Female":
    eixo.annotate("F",(df_root.iloc[i]["Annual Income (k$)"],df_root.iloc[i]["Spending Score (1-100)"]))
  else:
    eixo.annotate("M",(df_root.iloc[i]["Annual Income (k$)"],df_root.iloc[i]["Spending Score (1-100)"]))

plt.show()
```
