```python
df = pd.read_csv("shows.csv", sep=",")
```

```python
df.head(13)
```

### PRÉ-PROCESSAMENTO
#### TRATAMENTO DE ATRIBUTOS PARA VALORES NUMÉRICOS

```python
df["Nationality"] = df["Nationality"].map({"UK":0, "USA":1, "N":2})
df["Go"] = df["Go"].map({"NO":0, "YES":1})
df.head(13)
```

### SEPARAÇÃO DO DATASET EM ATRIBUTOS E ALVO


```python
X = df.iloc[:,:4]   # atributos
y = df.iloc[:,4]    # alvo
```

### MODELO DE MACHINE LEARNING

```python
modelo = DecisionTreeClassifier()
modelo.fit(X.values,y.values)
```

```python
plt.figure(figsize=(15,8))
plot_tree(modelo, filled=True, rounded=True, class_names=["Não", "Sim"], feature_names=["Age", "Experience", "Rank", "Nationality"])
plt.show()
```

### PREDIÇÃO

```python
print(modelo.predict([[40, 10, 7, 1]]))
```
