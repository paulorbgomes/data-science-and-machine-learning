Bibliotecas ...

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

Dataset ...

```python
arquivo = "shows.csv"
df = pd.read_csv(arquivo, sep=",")
df
```

Pré-Processamento ...

```python
df["Nationality"] = df["Nationality"].map({"UK":0, "USA":1, "N":2})
df["Go"] = df["Go"].map({"NO":0, "YES":1})
df
```

```python
attributes = ["Age", "Experience", "Rank", "Nationality"]
target = ["Go"]
X = df[attributes]
y = df[target]
```

Modelo de ML ...

```python
modelo = DecisionTreeClassifier()
modelo.fit(X.values,y.values)

plt.figure(figsize=(15,8))
plot_tree(modelo, filled=True, rounded=True, class_names=["Não", "Sim"], feature_names=["Age", "Experience", "Rank", "Nationality"])
plt.show()
```

Etapa de Predição ...

```python
print(modelo.predict([[36, 20, 10, 0]]))
```
